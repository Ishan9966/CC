```python
# Q1
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter the value of M: "))
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, 'b--', label="y = M * x^2")
plt.plot(x, y2, 'r-', label="y = M * sin(x)")
plt.title("Mathematical Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Q2
import pandas as pd
import seaborn as sns

subjects = ["Math", "Science", "English", "History", "Computer"]
scores = [85, 90, 78, 88, 95]
df = pd.DataFrame({"Subject": subjects, "Score": scores})

plt.figure(figsize=(8, 5))
sns.barplot(x="Subject", y="Score", data=df, palette="viridis")
for index, value in enumerate(scores):
    plt.text(index, value + 1, str(value), ha="center", fontsize=12)
plt.title("Scores in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y')
plt.show()

# Q3
np.random.seed(102316074) 
data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), 'g-')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)

axes[0, 1].scatter(range(50), data, c='b', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Values")
axes[0, 1].grid(True)

plt.tight_layout()
plt.show()

# Q4
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")

plt.figure(figsize=(10, 5))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker='o', color='r')
plt.title("Total Profit of All Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
for col in df.columns[1:]:
    sns.lineplot(x=df["month_number"], y=df[col], label=col)
plt.title("Product Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
df.set_index("month_number").plot(kind="bar", figsize=(12, 6))
plt.title("Sales Data of All Features")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
